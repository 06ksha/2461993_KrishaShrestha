# Problem 2: Advanced HDI Exploration

```python
#Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#Loading full dataset
file_path = "/content/drive/MyDrive/240085_KRISHASHRESTHA_ASSESMENT1/HDI_problem1B.csv"
hdi_df = pd.read_csv(file_path)
```

```python
# Ensuring numeric columns
numeric_cols = ['hdi', 'life_expectancy', 'gross_inc_percap', 'gender_development']
for col in numeric_cols:
    hdi_df[col] = pd.to_numeric(hdi_df[col], errors='coerce')
```

```python
# Task 1
# Creating South Asia Subset
south_asia_countries = ["Afghanistan", "Bangladesh", "Bhutan", "India",
                        "Maldives", "Nepal", "Pakistan", "Sri Lanka"]

# Filtering dataset for South Asia
sa_df = hdi_df[hdi_df['country'].isin(south_asia_countries)]
```

```python
# Saving the subset to CSV
sa_df.to_csv("/content/drive/MyDrive/240085_KRISHASHRESTHA_ASSESMENT1/HDI_SouthAsia.csv", index=False)

print("South Asia subset saved as HDI_SouthAsia.csv")
```

```python
# Normalizing Life Expectancy
sa_df.loc[:, 'life_expectancy_index'] = (sa_df['life_expectancy'] - sa_df['life_expectancy'].min()) / \
                                        (sa_df['life_expectancy'].max() - sa_df['life_expectancy'].min())

# Normalizing GNI per Capita
sa_df.loc[:, 'gni_percap_index'] = (sa_df['gross_inc_percap'] - sa_df['gross_inc_percap'].min()) / \
                                   (sa_df['gross_inc_percap'].max() - sa_df['gross_inc_percap'].min())
```

```python
#Task 2
#Composite Score
sa_df.loc[:, 'Composite_Score'] = 0.30 * sa_df['life_expectancy_index'] + 0.30 * sa_df['gni_percap_index']

# Ranking countries
sa_df = sa_df.sort_values(by='Composite_Score', ascending=False)
```

```python
# Ranking countries by Composite Score
sa_ranked = sa_df.groupby('country')['Composite_Score'].mean().sort_values(ascending=False)

# Converting to DataFrame for cleaner print
sa_ranked_df = sa_ranked.reset_index()

print("\nComposite Score Ranking (South Asia):")
print(sa_ranked_df.to_string(index=False))
```

```python
# Ranking countries by HDI
sa_hdi_ranked = sa_df.groupby('country')['hdi'].mean().sort_values(ascending=False)

# Converting to DataFrame for clean printing
sa_hdi_ranked_df = sa_hdi_ranked.reset_index()

print("\nHDI Ranking (South Asia):")
print(sa_hdi_ranked_df.to_string(index=False))
```

```python
# Top 5 countries by Composite Score
top5_composite_df = sa_ranked.head(5).reset_index()
top5_composite_df.columns = ['country', 'Composite_Score']

plt.figure(figsize=(8,5))

sns.barplot(
    x='Composite_Score',
    y='country',
    data=top5_composite_df,
    hue='country',
    dodge=False,
    palette='viridis',
    legend=False
)

plt.title("Top 5 South Asian Countries by Composite Score")
plt.xlabel("Composite Score")
plt.ylabel("Country")
plt.gca().invert_yaxis()

# Adding numeric labels
for i, v in enumerate(top5_composite_df['Composite_Score']):
    plt.text(v + 0.01, i, f"{v:.3f}", va='center')

plt.show()
```

```python
#Task 3
# Outlier Detection (Using 1.5 x IQR)
def detect_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR
    return df[(df[col] < lower) | (df[col] > upper)]
```

```python
# HDI outliers
hdi_outliers = detect_outliers(sa_df, 'hdi')
print("\nHDI Outliers:")
print(hdi_outliers[['country','hdi']])
```

```python
# GNI per capita outliers
gni_outliers = detect_outliers(sa_df, 'gross_inc_percap')

```

```python
# Scattering plot: GNI vs HDI with outliers highlighted
plt.figure(figsize=(8,6))
sns.scatterplot(data=sa_df, x='gross_inc_percap', y='hdi', label='Normal')
sns.scatterplot(data=gni_outliers, x='gross_inc_percap', y='hdi', color='red', label='Outliers')
plt.title("GNI per Capita vs HDI (South Asia)")
plt.xlabel("GNI per Capita")
plt.ylabel("HDI")
plt.legend()
plt.show()
```

```python
#Task 4
#  Exploring Metric Relationships
metrics = ['gender_development','life_expectancy']
for metric in metrics:
    correlation = sa_df[[metric,'hdi']].corr().iloc[0,1]
    print(f"\nCorrelation of {metric} with HDI: {correlation:.3f}")

    # Scatter plot with trendline
    plt.figure(figsize=(8,5))
    sns.regplot(data=sa_df, x=metric, y='hdi')
    plt.title(f"{metric.replace('_',' ').title()} vs HDI (South Asia)")
    plt.show()
```

```python
#Task 5
# GNI-HDI Gap Analysis
sa_df = sa_df.copy()  # first make a copy to avoid modifying the slice

sa_df.loc[:, 'GNI_HDI_Gap'] = sa_df['gross_inc_percap'] - sa_df['hdi']
```

```python
#Task 7
#Function to detect outliers using 1.5 x IQR
def detect_outliers(df, column, factor=1.5):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - factor * IQR
    upper = Q3 + factor * IQR
    return df[(df[column] < lower) | (df[column] > upper)]
# Function to handle outliers and plotting for a region ---
def hdi_outlier_plot(region_name, df):
    # Detect HDI outliers
    hdi_out = detect_outliers(df, 'hdi')
    if hdi_out.empty:
        lowest_country = df.groupby('country')['hdi'].mean().sort_values().head(1).index
        hdi_out = df[df['country'].isin(lowest_country)]

    # Scatter plot
    plt.figure(figsize=(8,6))
    plt.scatter(df['gross_inc_percap'], df['hdi'], color='skyblue', label='Normal')
    plt.scatter(hdi_out['gross_inc_percap'], hdi_out['hdi'],
                color='red', s=100, label='HDI Outlier', edgecolor='black')

    plt.xlabel("GNI per Capita")
    plt.ylabel("HDI")
    plt.title(f"{region_name}: HDI vs GNI per Capita with HDI Outliers")
    plt.legend()
    plt.grid(True)
    plt.show()

    # Printing HDI outlier table
    print(f"{region_name} – HDI Outliers:")
    print(hdi_out[['country','year','hdi']])
    print("\n" + "-"*60 + "\n")

# South Asia
hdi_outlier_plot("South Asia", sa_df)
# Middle East
hdi_outlier_plot("Middle East", me_df)
```
